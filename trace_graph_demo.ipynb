```python
from micrograd.engine import Value
from drawgraph.neuralnet import draw_dot
```

```python
def neuron(x1_val, x2_val, w1_val, w2_val, b_val):
    x1 = Value(x1_val, label="x1")
    x2 = Value(x2_val, label="x2")
    w1 = Value(w1_val, label="w1")
    w2 = Value(w2_val, label="w2")
    b = Value(b_val, label="b")

    x1w1 = x1 * w1
    x1w1.label = "x1*w1"

    x2w2 = x2 * w2
    x2w2.label = "x2*w2"

    x1w1x2w2 = x1w1 + x2w2
    x1w1x2w2.label = "x1w1+x2w2"

    n = x1w1x2w2 + b
    n.label = "n"

    o = n.tanh()
    o.label = "o"

    o.backward()
    return o
```

```python
# Run the neuron computation and visualize
output = neuron(2.0, 0.0, -3.0, 1.0, 6.881373587)
draw_dot(output)
```
